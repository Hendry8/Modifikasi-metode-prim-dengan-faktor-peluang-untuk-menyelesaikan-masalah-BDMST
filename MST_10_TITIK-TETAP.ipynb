```python
def membuat_tabel(n): #membuat himpunan1
    global sumber1,df
    sumber1 =[]

    i = 0
    while i < len(n):    
        for j in range(len(n)-1,0,-1):

            k=0
            while k < j:
                sumber1.append(n[i])

                k+=1
            i+=1   
            if len(sumber1)==int(len(n)*(len(n)-1)/2):
                break
        if len(sumber1)==int(len(n)*(len(n)-1)/2):
            break

    global tujuan1
    tujuan1 = []
    for i in range(1,len(n)):
        a = n[i:len(n)]
        for j in range(len(a)):
            tujuan1.append(a[j])
    
    
    data= {"sumber":sumber1,
            "tujuan":tujuan1,
              "bobot":df}
    df = pd.DataFrame(data)
```

```python
def awal(df): #mendeklarasikan semua vortex dalam graf
    global vortex,graf,root
    vortex = []
    
    for i in range(len(df)):
        if df.iloc[i]['tujuan'] not in vortex:
            vortex.append(df.iloc[i]['tujuan'])
        if df.iloc[i]['sumber'] not in vortex:
            vortex.append(df.iloc[i]['sumber'])
    vortex
    #membuar akar random dari Graph
    
    root= 1 #mencari titik awal dengan random 
    graf=pd.DataFrame( columns = ['sumber', 'tujuan', 'bobot'])
    for i in range(len(df)):
        if df.iloc[i]['tujuan'] == root:
            graf = graf.append(df.iloc[i])
        if df.iloc[i]['sumber'] == root:
            graf = graf.append(df.iloc[i])
    graf=graf.sort_values(by=['bobot']) #mengambil sisi paling minimum pada root 
    graf = graf.iloc[[0]]
```
